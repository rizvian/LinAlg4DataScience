```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([4,5,1])
b = np.array([-4,0,-4])
c = np.array([1,3,2])
print ((1*a)+(2*b)+(-3*c))
print(2*b)
```

```python
a = np.array([5,3])
b = np.array([7,0])
lc = np.dot(b,a)/np.dot(b,b)
c = b*lc
d = a - c

print("a=",a,"b=",b,"c=",c, "lc=",lc)

plt.arrow(0,0,a[0],a[1],head_width=0.5, length_includes_head=True)
plt.arrow(0,0,b[0],b[1],head_width=0.5, length_includes_head=True)
plt.arrow(a[0],a[1], c[0]-a[0], c[1]-a[1], head_width=0.5, length_includes_head=True)
plt.arrow(0,0,d[0],d[1],head_width=0.5, length_includes_head=True)
plt.axis([-10,10,-10,10])
plt.grid()
```
