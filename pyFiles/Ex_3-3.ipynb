```python
import plotly.graph_objects as go
import numpy as np 
import random as rnd
b=np.array([[1,3]])
s=np.array([[rnd.uniform(1,10) for x in range(1,100)]])
v=s.T*b
fig = go.Figure(go.Scatter(x= v[:,0], y= v[:,1], mode="markers", marker=dict(size=4)))
fig.update_layout(xaxis=dict(range=[0,50], dtick=2), yaxis=dict(range=[0,50],dtick=2), yaxis_scaleanchor="x", height=500, width=500)
fig.show()
```

```python
b=np.array([[3,5]])
s1=np.array([[rnd.uniform(1,10) for x in range(1,100)]])
s2=np.array([[rnd.uniform(1,10) for x in range(1,100)]])
v1, v2=s1.T*b, s2.T*b
fig = go.Figure(go.Scatter3d(x= v1[:,0], y= v1[:,1], z=v2[:,1], mode="markers", marker=dict(size=4)))
fig.update_layout(xaxis=dict(range=[0,10], dtick=1), yaxis=dict(range=[0,10],dtick=1), yaxis_scaleanchor="x", height=500, width=500)
fig.show()
```
